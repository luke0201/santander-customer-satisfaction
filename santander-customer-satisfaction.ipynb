# Santander Customer Satisfaction

## Data Preparation

```python
import pandas as pd

df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

Separate the features and the label.

```python
y = df['TARGET']
X = df.drop('TARGET', axis=1)
```

## Data Wrangling

It seems that *ID* would not have any relationship with our later predictions. Hence, drop it.

```python
X.drop('ID', axis=1, inplace=True)
```

```python
X.describe()
```

In *var3* column, -999999 is too far from other values. Replace it with the median from *var3*, which is 2.

```python
X['var3'].replace(-999999, 2, inplace=True)
```

Split the dataset into train and test parts.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Train Model

We are going to use LightGBM classifier. Hyperparameters are determined by grid search.

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'num_leaves': [32, 64],
    'max_depth': [128, 160],
    'min_child_samples': [60, 100],
    'subsample': [0.8, 1]
}
lgbm_classifier = LGBMClassifier(n_estimators=200)
grid_search_cv = GridSearchCV(lgbm_classifier, param_grid=param_grid, cv=3)

grid_search_cv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric='auc', eval_set=[(X_train, y_train)])

estimator = grid_search_cv.best_estimator_
```

## Test Model

```python
from sklearn.metrics import roc_auc_score

y_score = estimator.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_score, average='macro')
print(f'ROC AUC on the test dataset: {roc_auc:.4f}')
```
